```python

def preprocess(input_text):
    # Convert the text to lowercase
    text = input_text.lower()

    # Expand contractions
    pattern = re.compile(r'\b(' + '|'.join(contraction_dict.keys()) + r')\b')

    def replace(match):
        return contraction_dict[match.group(0)]

    # Use re.sub with the pattern to replace contractions with their expansions
    text = pattern.sub(replace, text)

    # Remove punctuation
    text = ''.join(char for char in text if char not in string.punctuation)

    # Remove stop words
    stop_words = set(nltk.corpus.stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]

    # Join the filtered words back into a single string
    text = ' '.join(filtered_words)

    return text

```

```python
def ngram_tokenization(text_data, ngram_range=(1, 2)):
      vectorizer = CountVectorizer(ngram_range=ngram_range)
      X = vectorizer.fit_transform(text_data)
      feature_names = vectorizer.get_feature_names_out()
      X_array = X.toarray()

      return feature_names, X_array


```

```python
data['processed_likes']=data['Likes'].apply(preprocess)
data['processed_dislikes']=data['Dislikes'].apply(preprocess)

```

```python
data['processed_dislikes']
```

```python
# Tokenize the processed "Likes" column
feature_names_likes, tokenized_likes = ngram_tokenization(data['processed_likes'])
# Tokenize the processed "Dislikes" column
feature_names_dislikes, tokenized_dislikes = ngram_tokenization(data['processed_dislikes'])
```

```python
# Define the number of topics
num_topics = 5  # You can adjust this number based on your dataset

# Create an LDA model for Likes
lda_likes = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_likes.fit(tokenized_likes)

# Create an LDA model for Dislikes
lda_dislikes = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_dislikes.fit(tokenized_dislikes)

# Extract and print the top keywords for each topic
def print_top_keywords(model, feature_names, n_words=10):
    for topic_idx, topic in enumerate(model.components_):
        top_keywords_idx = topic.argsort()[-n_words:][::-1]
        top_keywords = [feature_names[i] for i in top_keywords_idx]
        print(f"Topic {topic_idx + 1}: {', '.join(top_keywords)}")

# Print top keywords for Likes
print("Top Keywords for Likes:")
print_top_keywords(lda_likes, feature_names_likes)

# Print top keywords for Dislikes
print("\nTop Keywords for Dislikes:")
print_top_keywords(lda_dislikes, feature_names_dislikes)
```
